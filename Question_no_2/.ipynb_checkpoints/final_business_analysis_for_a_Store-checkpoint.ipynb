```python
import pandas as pd
import numpy as np 
from datetime import datetime
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6
```

```python
join_type = ['inner']
features_data = pd.read_csv("Data_set/Features data set.csv")
sales_data = pd.read_csv("Data_set/sales data-set.csv")
```

```python
features_data.head()
```

```python
sales_data.head()
```

```python
weekly_sales_data = pd.merge(features_data, sales_data, how=join_type[0], on=['Date', 'Store', 'IsHoliday'])
```

```python
weekly_sales_data['Date'] = pd.to_datetime(weekly_sales_data['Date'])
weekly_sales_data['Date'] = weekly_sales_data['Date'].dt.date
```

```python
weekly_sales_data.head()
```

```python
# TOTAL WEEKLY MARKDOWN
markdown_columns = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
weekly_sales_data[markdown_columns] = weekly_sales_data[markdown_columns].fillna(value=0)
weekly_sales_data['MarkDown'] = weekly_sales_data['MarkDown1'] + weekly_sales_data['MarkDown2'] + \
weekly_sales_data['MarkDown3'] + weekly_sales_data['MarkDown4'] + weekly_sales_data['MarkDown5']
```

```python
weekly_sales_data.head()
```

```python
# Let's keep only required columns
kept_columns = ['Store', 'Date', 'Temperature', 'Fuel_Price', 'MarkDown', 'CPI', 'Unemployment', 'IsHoliday', 'Weekly_Sales']
weekly_sales_data = weekly_sales_data[kept_columns]
```

```python
weekly_sales_data.head()
```

```python
# Keeping data for only on column
# Here let's take different Store and Plot correlation between between different attributes 
# and weekly sales
weekly_sales_data = weekly_sales_data[weekly_sales_data.Store==1]
```

```python
len(weekly_sales_data)
```

```python
weekly_sales_data.tail()
```

```python
variables = ['MarkDown', 'Weekly_Sales']
markdown_weekly_sales_df = weekly_sales_data[variables]
```

```python
markdown_weekly_sales_corr = markdown_weekly_sales_df.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(markdown_weekly_sales_corr, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,2,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(variables)
ax.set_yticklabels(variables)
plt.show()
```
